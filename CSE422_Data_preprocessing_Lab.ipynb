```python
import numpy as np
import pandas as pd
import sklearn
```

**Loading the dataset as dataframe using pandas**

```python
dataset = pd.read_csv('/content/breast cancer classification dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop(['Unnamed: 32'], axis=1)
dataset.shape
```

```python
dataset = dataset.dropna(axis=0, subset=['radius_mean', 'fractal_dimension_worst'])

dataset.shape
```

```python
dataset.info()
```

```python
dataset['diagnosis'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

dataset['diagnosis'] = enc.fit_transform(dataset['diagnosis'])

print(dataset[['diagnosis']])
```

```python
from sklearn.model_selection import train_test_split

X = dataset.drop('diagnosis', axis = 1)
Y = dataset['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = 1)

print(X_train.shape)
print(X_test.shape)

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
```

```python
print(X_train.min(axis=0))
print('---------------')
print(X_train.max(axis=0))
```

```python
print(X_train_scaled.min(axis=0))
print('---------------')
print(X_train_scaled.max(axis=0))
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train_scaled, y_train)

print(knn.score(X_test_scaled, y_test))
```

```python
knn.fit(X_train, y_train)

print(knn.score(X_test, y_test))
```
